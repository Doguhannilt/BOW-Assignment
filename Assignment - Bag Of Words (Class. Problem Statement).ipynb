```python
df["category"] = df["sentiment"].apply(lambda x: 1 if x == "positive" else 0)
df["category"]
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')

df['review_without_stopwords'] = df['review'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
    
```

```python
df['review_without_stopwords'] = df['review_without_stopwords'].str.replace(r"\W", " ", regex=True)
df
```

```python
df['review_without_stopwords'] =  df['review_without_stopwords'].apply(lambda x: x.lower())
df.drop("review",axis=1,inplace=True)
df
```

```python
import nltk

w_tokenizer = nltk.tokenize.WhitespaceTokenizer()
lemmatizer = nltk.stem.WordNetLemmatizer()

def lemmatize_text(text):
    return [lemmatizer.lemmatize(w) for w in w_tokenizer.tokenize(text)]

df['review_without_stopwords'] = df.review_without_stopwords.apply(lemmatize_text)
```

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Set the desired sample size
sample_size = 1800

# Take a random sample from your dataframe
sample_data = df.sample(n=sample_size, random_state=42)

# Convert the preprocessed text column to a list of strings
text_list = [' '.join(row) for row in sample_data["review_without_stopwords"]]

# Vectorization
vectorizer = CountVectorizer()
vectorized_text = vectorizer.fit_transform(text_list)

# Convert the vectorized text to a DataFrame
vectorized_df = pd.DataFrame(vectorized_text.toarray(), columns=vectorizer.get_feature_names())
```

```python
X = vectorized_df
y = df["category"]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python

# Modelling
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
param_dist = {'n_estimators': randint(50,500),
              'max_depth': randint(1,20)}

# Create a random forest classifier
rf = RandomForestClassifier()

# Use random search to find the best hyperparameters
rand_search = RandomizedSearchCV(rf, 
                                 param_distributions = param_dist, 
                                 n_iter=5, 
                                 cv=5)

# Fit the random search object to the data
rand_search.fit(X_train, y_train)
```

```python
# Create a variable for the best model
best_rf = rand_search.best_estimator_

# Print the best hyperparameters
print('Best hyperparameters:',  rand_search.best_params_)
```

```python
y_pred = best_rf.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
from sklearn.naive_bayes import GaussianNB

# Build a Gaussian Classifier
model = GaussianNB()

# Model training
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    ConfusionMatrixDisplay,
    f1_score,
)

y_pred = model.predict(X_test)
accuray = accuracy_score(y_pred, y_test)
f1 = f1_score(y_pred, y_test, average="weighted")

print("Accuracy:", accuray)
print("F1 Score:", f1)
```

```python
from sklearn.model_selection import RepeatedStratifiedKFold

cv_method = RepeatedStratifiedKFold(n_splits=5, 
                                    n_repeats=3, 
                                    random_state=999)
```

```python
from sklearn.preprocessing import PowerTransformer
from sklearn.model_selection import GridSearchCV
params_NB = {'var_smoothing': np.logspace(0,-9, num=100)}

gs_NB = GridSearchCV(estimator=model, 
                     param_grid=params_NB, 
                     cv=cv_method,
                     verbose=1, 
                     scoring='accuracy')

Data_transformed = PowerTransformer().fit_transform(X_test)

gs_NB.fit(Data_transformed, y_test);
```
